### BLEU 和 ROUGE

BLEU：看模型输出里有多少“正确的词 / n-gram”
image.png

```python
from nltk.translate.bleu_score import sentence_bleu, SmoothingFunction

# 参考答案
reference = ["I love eating apples".split()]

# 模型输出
candidate_A = "I love cooking apples".split()
candidate_B = "Eating apples is something I really love".split()

smooth = SmoothingFunction().method1

bleu_A = sentence_bleu(reference, candidate_A, smoothing_function=smooth)
bleu_B = sentence_bleu(reference, candidate_B, smoothing_function=smooth)

print(f"BLEU - Model A: {bleu_A:.4f}")   # A会比较高，因为和参考答案的词语重叠最多（I，love，apples），虽然意思发生了一些改变但是blue计算不看这些
print(f"BLEU - Model B: {bleu_B:.4f}")  # B比较低，因为与原句重叠不多，虽然意思比A更接近，但是blue分数低

```

### rouge

image.png

```python
from rouge_score import rouge_scorer

ref = "I love eating apples"
cand_A = "I like cooking apples"
cand_B = "Eating apples is something I really love"

scorer = rouge_scorer.RougeScorer(["rouge1", "rougeL"], use_stemmer=True)

score_A = scorer.score(ref, cand_A)
score_B = scorer.score(ref, cand_B)

print("===== ROUGE - Model A =====")
print(score_A)
print("\n===== ROUGE - Model B =====")
print(score_B)
```

### Perplexity（困惑度，PPL）
image.png

```python
import math, torch
from transformers import AutoTokenizer, AutoModelForCausalLM

model_name = "gpt2"  # for English;
tok = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForCausalLM.from_pretrained(model_name)
model.eval()

def ppl(sentence: str) -> float:
    ids = tok(sentence, return_tensors="pt").input_ids
    with torch.no_grad():
        loss = model(ids, labels=ids).loss
    return math.exp(loss.item())

s1 = "The cat sat on the mat."
s2 = "Mat the on sat cat the."
s3 = "The cat sat on mat the ."
print("PPL S1:", ppl(s1))
print("PPL S2:", ppl(s2))  # 分数应该比S1高，因为并非人类的表达习惯
print("PPL S3:", ppl(s3))  # 分数应该比S2低，因为相对正常

```

### Exact Match（完全匹配，EM）

image.png

```python
import re

def normalize(s: str) -> str:
    s = s.lower().strip()
    # keep letters/numbers/space; drop punctuation
    s = re.sub(r"[^a-z0-9\s]", "", s)
    s = re.sub(r"\b(the|a|an)\b", "", s)  # drop English articles
    s = re.sub(r"\s+", " ", s).strip()
    return s

def exact_match(pred: str, ref: str) -> int:
    return int(normalize(pred) == normalize(ref))


ref = "Eiffel Tower"
pred1 = "the eiffel tower"
pred2 = "Eiffel Tower, Paris"

print("EM(pred1, ref):", exact_match(pred1, ref))  # 1，因为正则化之后和标准答案一样
print("EM(pred2, ref):", exact_match(pred2, ref))  # 0，因为多了额外的Paris输出，和标准答案出现了区别

```

### 事实正确率（Fact Accuracy / Factual Score）

与幻觉率相反，统计“回答正确的比例”
image.png

```python
def factual_accuracy(preds, refs):
    correct = sum(p.lower().strip() == r.lower().strip() for p, r in zip(preds, refs))
    return correct / len(preds)

refs = ["Paris", "Mount Everest"]
preds = ["Paris", "K2"]  # 只有第二个错误，所以为0.5
print("Fact Accuracy1:", factual_accuracy(preds, refs))

refs = ["Paris", "Mount Everest"]
preds = ["london", "K2"]  # 全错，应该为0
print("Fact Accuracy2:", factual_accuracy(preds, refs))
```

### 幻觉率（Hallucination Rate）

定义：模型回答中包含明显错误或不存在的事实的比例。
常用于：问答、RAG、知识密集任务。

两句对比示例

✅ 正确：

Question: Who discovered America?
Answer: Christopher Columbus.

❌ 幻觉：

Answer: America was discovered by Elon Musk in 1850.

```python
## 一个非常简单的幻觉检测，一旦与标签不一致则检测为幻觉，但是现实中会有别的方法来判断。
ground_truth = {
    "Who discovered America?": "Christopher Columbus",
    "Capital of France?": "Paris",
    "Capital of China?": "Beijing"
}

preds = [
    "Christopher Columbus",
    "The capital of France is London",
    "Beijing"
]

def hallucination_rate(questions, preds, gt):
    hallucinations = 0
    for q, p in zip(questions, preds):
        if gt[q].lower().strip() != p.lower().strip():
            hallucinations += 1
    return hallucinations / len(preds)

qs = list(ground_truth.keys())
print("Hallucination Rate:", hallucination_rate(qs, preds, ground_truth))  # 3个回答有一个是错误的，因此幻觉率为1/3
```

### 稳定性 / 一致性（Consistency）

如果同一个问题，多次回答都不一样 → 模型不稳定

✅ 两句对比示例

问三次：

Answer: "Paris" ✅

Answer: "Paris" ✅

Answer: "London" ❌（不一致）

```python
from collections import Counter

def consistency_score(answers):
    most_common_count = Counter(answers).most_common(1)[0][1]
    return most_common_count / len(answers)

answers = ["Paris", "Paris", "London"]
print("Consistency:", consistency_score(answers))  # 2/3 = 0.66
```

### Latency（延迟）

定义：一次请求从发出 → 收到第一 token → 收到最后 token所需时间
通常分：

TTFT：Time To First Token

TTLT / TTFB：Time To Last Token / Time To Full Buffer

```python
import time
from transformers import AutoTokenizer, AutoModelForCausalLM
import torch

model_name = "gpt2"
tok = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForCausalLM.from_pretrained(model_name)

prompt1 = "Explain quantum physics in one sentence."
prompt2 = "Hi."
input1 = tok(prompt1, return_tensors="pt")
input2 = tok(prompt2, return_tensors="pt")

start1 = time.time()
with torch.no_grad():
    out1 = model.generate(**input1, max_new_tokens=50)
end1 = time.time()

start2 = time.time()
with torch.no_grad():
    out1 = model.generate(**input2, max_new_tokens=50)
end2 = time.time()

print("Latency (s):", end1 - start1)  # 看一次请求需要多长的时间
print("Latency (s):", end1 - start1)
```

### Throughput（QPS：每秒处理请求数）

定义：每秒能同时处理多少次请求

两句对比

模型 A：每秒 100 个请求 ✅

模型 B：每秒 5 个请求 ❌ → 不适合线上产品

```python
import time
from transformers import AutoTokenizer, AutoModelForCausalLM
import torch

model_name = "gpt2"
tok = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForCausalLM.from_pretrained(model_name)

prompts = ["Hello!"] * 10  # 10个请求模拟，最后用10/总时间，即可得到每秒多少个请求
inputs = tok(prompts, return_tensors="pt")
start = time.time()

with torch.no_grad():
    _ = model.generate(**inputs, max_new_tokens=20)

end = time.time()
qps = len(prompts) / (end - start)
print("QPS:", qps)
```

### Tokens per Second（令牌生成速度）

更贴近用户体验。

两句对比

模型 A：30 tokens/s ✅ 很快

模型 B：2 tokens/s ❌ 输出卡顿

```python
import time
from transformers import AutoTokenizer, AutoModelForCausalLM

tok = AutoTokenizer.from_pretrained("gpt2")
model = AutoModelForCausalLM.from_pretrained("gpt2")

prompt = "Write a short story about a robot."
inputs = tok(prompt, return_tensors="pt")

start = time.time()
with torch.no_grad():
    out = model.generate(**inputs, max_new_tokens=100)
end = time.time()

num_tokens = out.shape[-1] - inputs.input_ids.shape[-1]
print("Tokens/sec:", num_tokens / (end - start))
```
